**Alterando o tipo de dado da 42 colunas que se referem a valores numericos, para assim trabalhar melhor com os dados que possuo.**

```python
colunas = df.columns[3:45]
for columns in colunas:
  df["{}".format(columns)] = pd.to_numeric(df["{}".format(columns)], errors='coerce')
```

**Nessa linha de código estou verificando a quantidade de informações que tenho referente a cada feature encontrada na tabela.**
